```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lr
```

```python
pd.options.display.float_format = "{:,.2f}".format
```

```python
data = pd.read_csv('./data/cost_revenue_dirty.csv')
data[['production_budget_usd', 'worldwide_gross_usd']].describe()
```

```python
data
```

```python
x = pd.DataFrame(data, columns=['production_budget_usd'])
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y, alpha=0.3)
plt.title('Film Cost vs Revenue')
plt.xlabel('Budger USD')
plt.ylabel('Revenue USD')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

## Linear Regression
Linear regression is a statistical method used to find the relationship between a dependent variable and one or more independent variables. It assumes that there is a linear relationship between the input variables and the output variable.

[IBM Linear Regression](https://www.ibm.com/topics/linear-regression)

```python
regression = lr.LinearRegression()
regression.fit(x, y)
```

Regression Slope Coeficient and Intercept

```python
(regression.coef_, regression.intercept_)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y, alpha=0.3)
plt.plot(x, regression.predict(x), color='red')
plt.title('Film Cost vs Revenue')
plt.xlabel('Budger USD')
plt.ylabel('Revenue USD')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

```python
regression.score(x, y)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Create some sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Create a LinearRegression object and fit the data
reg = LinearRegression().fit(X, y)

# Predict using the fitted model
y_pred = reg.predict([[6], [7]])

# Print the coefficients and the predicted values
print("Coefficients:", reg.coef_)
print("Intercept:", reg.intercept_)
print("Predicted values:", y_pred)
```

In this example, we first create some sample data with one independent variable (X) and one dependent variable (y). We then create a LinearRegression object and fit the data using the `fit()` method. This calculates the coefficients for the linear equation that best fits the data. We can then use the `predict()` method to predict the values for new data points (`[[6], [7]]` in this case).

Finally, we print out the coefficients (slope) and intercept of the linear regression model, as well as the predicted values for the new data points.

This means that the linear equation that best fits the data is `y = 0.6x + 1.8`, where x is the independent variable and y is the dependent variable. The predicted values for the new data points are `5.4` and `6.0`
