```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
years = [x for x in range(2010, 2021)]
temps = [0.72,0.61,0.65,0.68,0.75,0.90,1.02,0.93,0.85,0.99,1.02]
```

```python
plt.plot(years, temps, marker = '^')
plt.xlabel("Year")
plt.ylabel("Temperatures")
plt.title('Global Warming',{'fontsize':12,'horizontalalignment':'center'})
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] 
customer1 = [12,13,9,8,7,8,8,7,6,5,8,10]
customer2 = [14,16,11,7,6,6,7,6,5,8,9,12]
customer3 = np.random.randint(8, 15, 12)

plt.plot(months, customer1, label='Customer 1', marker = 'o')
plt.plot(months, customer2, label='Customer 2', marker = '^')
plt.plot(months, customer3, label='Customer 3', marker = 's', color='limegreen')
plt.xlabel("Months")
plt.ylabel("Customers")
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(months, customer2, label = 'Customer2', marker = '^')
plt.xlabel('Month')
plt.title('Building Consumption of Customer2')
plt.show()
```

```python
plt.scatter(months, customer1, label='Customer 1', marker = 'o')
plt.scatter(months, customer2, label='Customer 2', marker = '^')
plt.scatter(months, customer3, label='Customer 3', marker = 's', color='limegreen')
plt.xlabel("Months")
plt.ylabel("Customers")
plt.legend()
plt.show()
```

```python
plt.hist(customer1, bins = 20)
plt.xlabel("Months")
plt.ylabel('Elec Consumption')
plt.title('Histogram')
plt.show()
```

```python
bar_width = 0.4
Month_b = np.arange(12)
plt.bar(Month_b, customer1, bar_width, label = 'Customer1')
plt.bar(Month_b+bar_width, customer2, bar_width, label = 'Customer2')
plt.xlabel("Months")
plt.ylabel('Kw')
plt.title('Elec Consumption')
plt.legend()
plt.show()
```

```python
plt.boxplot([customer1, customer2], vert=False)
```
