### Series
A one-dimensional labeled array that can hold data of any type. It is similar to a column in a spreadsheet or a single variable.

```python
age = pd.Series([x * 10 for x in range(1, 10)], index=[f"age{x}" for x in range(1, 10)])
```

```python
print(age)
print(age['age2'])
for x in [3, 5, 7, 8]:
    print(f"age{x}: {age[f'age{x}']}")
```

```python
# Filters
print(age[age>30])
print(age[(age>30) & (age<80)])
```

### Dataframes
A two-dimensional labeled data structure with columns of potentially different data types. It resembles a table or a spreadsheet, with rows and columns.

```python
np_data = np.random.randint(1, 100, (10, 8))
data_frame = pd.DataFrame(np_data, columns=[f"col{x}" for x in range(1,9)])
print(data_frame)
# Alter dataset
data_frame['col9'] = pd.Series(np.random.randint(1, 200, 10))
print(data_frame)
data_frame.describe()
```

```python
print(data_frame.iloc[3])
print(data_frame.iloc[2:6])
print(data_frame.drop(columns=['col3']))
```
