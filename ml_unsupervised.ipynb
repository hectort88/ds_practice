```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
data_iris = iris.data # Only use the data, not the target labels
```

```python
# K-means clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3) # define 3 clusters
kmeans.fit(data_iris)

# Predict the labels
labels = kmeans.predict(data_iris)
centers = kmeans.cluster_centers_
```

```python
plt.style.use('ggplot')
plt.scatter(data_iris[:,2], data_iris[:,3], c=labels)
plt.scatter(centers[:,2], centers[:,3], marker='o', color='blue', s=120)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

```python
kmeans.inertia_
```

```python
# Best number of clusters
k_inertia = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=44) # define n clusters
    kmeans.fit(data_iris)
    k_inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), k_inertia,  marker='o')
plt.show()
```

```python
# DBSCAN: Density Based Spatial Clustering of Application with Noise
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=0.7, min_samples=4)
dbs.fit(data_iris)
dbs_labels = dbs.labels_
```

```python
plt.scatter(data_iris[:,2], data_iris[:,3], c=dbs_labels)
# plt.scatter(centers[:,2], centers[:,3], marker='o', color='blue', s=120)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
hr_linkage = linkage(data_iris, method='complete')
den = dendrogram(hr_linkage)
```

```python
hr_labes = fcluster(hr_linkage, t=4, criterion='distance')
```

```python
plt.scatter(data_iris[:,2], data_iris[:,3], c=hr_labes)
# plt.scatter(centers[:,2], centers[:,3], marker='o', color='blue', s=120)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
